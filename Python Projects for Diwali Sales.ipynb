Exploratory Data Analysis


```python
df.columns
```

```python
sns.countplot(x= 'Gender', data= df)
```

```python
ax = sns.countplot(x= 'Gender', data= df)

#to get the value of Bar
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#group by
df.groupby(['Gender'], as_index=False)['Amount'].sum()
```

```python
Sales_gen =df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
```

```python
sns.barplot(x='Gender',y='Amount',data= Sales_gen)
```

```python
#Without Hue
sns.countplot(data =df, x='Age Group')
```

```python
#With Hue
sns.countplot(data =df, x='Age Group', hue= "Gender")
```

```python
ax = sns.countplot(data =df, x='Age Group', hue= "Gender")

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#Total Amount vs Age Group
SalesAge = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group',y='Amount', data = SalesAge)
```

*Here We can see most of Buyer are between 26-35 Age Group*

```python
# Total Number of order in top 10 states
StateSale = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#figure Size
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data = StateSale)
```

```python
df.columns
```

```python
# Total Number of Amount in top 10 states
StateAmt = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)

#figure Size
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Amount', data = StateAmt)
```

Most of the Amonut/State comes form UP, Maharashtra & Karnataka

```python
ax= sns.countplot(x='Marital_Status', data = df)

sns.set(rc={'figure.figsize':(5,7)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
SalesGen = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data = SalesGen,x='Marital_Status', y='Amount',  hue = 'Gender')
```
